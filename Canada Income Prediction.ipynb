```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('years')
plt.ylabel('Incom')
plt.scatter(df.year,df.incom,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.incom)
```

```python
reg.predict([[2025]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
